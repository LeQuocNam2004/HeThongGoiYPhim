# Khám phá dữ liệu phim (Exploratory Data Analysis)

Notebook này được sử dụng để khám phá tập dữ liệu phim từ TMDB.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc dữ liệu đã qua tiền xử lý
movies = pd.read_pickle('../models/movies_clean.pkl')
movies.head()
```

## Phân tích điểm đánh giá (Vote Average)

```python
plt.figure(figsize=(10,6))
sns.histplot(movies['vote_average'], bins=30, kde=True)
plt.title('Phân phối điểm đánh giá phim')
plt.xlabel('Điểm đánh giá')
plt.ylabel('Số lượng phim')
plt.show()
```
